```python
import csv
data = csv.reader(open('price_1982_2021.csv', encoding = 'cp949'))
next(data)

x = []
year = list(range(1990, 2022))  #치킨의 가격은 1990년부터 나와있다.

for row in data:
  if '치킨' in row[1]:
    for i in range(10, len(row)):
      x.append(float(row[i])*180)   #원래 통계 파일에서는 2020년 치킨의 가격을 100으로 환산한 지수로 나오는데, 학생들에게 가격이 와닿게 하기 위해 2020년 전국 평균 치킨 가격을 임의로 18000원으로 설정. 각 년도의 지수에 180을 곱하였다.


import matplotlib.pyplot as plt
plt.plot(year, x, label='치킨의 가격 변화')
plt.legend()
plt.title('1990년부터 2021년까지 치킨의 가격 변화')
plt.xlabel('년도')
plt.ylabel('가격')
plt.show()
      



```

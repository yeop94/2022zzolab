## 내 생일을 입력하고, 내 생일날의 구미시 최고기온과 최저기온을 그래프로 알아보기
###1) 경상북도 구미시 1973년~2021년까지의 기온 데이터
###2) 생일 입력하기
###3) 년도별 최고기온과 최저기온의 변화를 그래프로 알아보기

```python
import numpy as np
import time
import datetime as dt
import csv
data = csv.reader(open('gumitemp.csv', encoding = 'cp949'))  # 구미시 기온 데이터 업로드

while True:
  bday = input('당신의 생일을 입력해주세요 [예: 09-28] : ')
  try:
    dt.datetime.strptime(bday, "%m-%d")                 # str인 값을 datetime으로 변경
  except ValueError:                                    # 날짜 형태가 틀리면
    print('00-00 형태로 다시 입력해주세요')
    continue
  else:
    break

year = []
hightemp = []
lowtemp = []

for row in data:
  if row[-1] and row[-2] != '':
    if '-' + bday in row[0]:
      year.append(int(row[0][0:4]))
      hightemp.append(float(row[-1]))
      lowtemp.append(float(row[-2]))

import matplotlib.pyplot as plt
plt.plot(year, hightemp, label='the highest temperature')
plt.plot(year, lowtemp, label='the lowest temperature')
plt.legend()
plt.xlabel('year')
plt.ylabel('temperature')
plt.show()
```
